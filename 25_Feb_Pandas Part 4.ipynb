```python
df=pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
```

```python
df
```

```python
# Here the fare is given in Euros so convert into the rupees
# Apply function comes into the play if you have to perform a single function into the entire column
df['Fare_INR']=df['Fare'].apply(lambda x:x*89.04)
# This is using a inside function
```

```python
df.head()
```

```python
# Same using the outside Function
def euro_inr(x):
    return x*89.04
df['Fare_INR']=df['Fare'].apply(euro_inr)
```

```python
df.head()
```

```python
# To count the length of the name i.e Number of haracters in the name 
df['Name_len']=df['Name'].apply(len)
```

```python
df
```

```python
# If the fare is less than 10 Euro the write cheap mid for between 10 and 20 and Highly priced for more than 20
def cat_fare(x):
    if x<10:
        return 'Cheap'
    elif x>=10 and x<20:
        return 'Mid'
    else:
        return 'Highly Priced'
# Function is created now calling it

df['Fare_Category']=df['Fare'].apply(cat_fare)
```
