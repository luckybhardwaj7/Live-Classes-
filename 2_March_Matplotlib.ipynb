### Scatter Plot

```python
x=np.random.rand(50)
y=np.random.rand(50)
plt.scatter(x,y)
# This is distributed or Scattered Data forming no Relationship 
```

```python
x=np.random.rand(50)
y=np.random.rand(50)
colours=np.random.rand(50) # For different colours
plt.scatter(x,y,c=colours)
```

```python
x=np.random.rand(50)
y=np.random.rand(50)
sizes=200*np.random.rand(50) # *200 is just to increase the size of the dots you can change it depending on your need
colours=np.random.rand(50) # For different colours
plt.scatter(x,y,c=colours,s=sizes)
```

```python
x=np.random.rand(50)
y=np.random.rand(50)
sizes=200*np.random.rand(50) # *200 is just to increase the size of the dots you can change it depending on your need
colours=np.random.rand(50) # For different colours
plt.scatter(x,y,c=colours,s=sizes,alpha=1) # Alpha is for intensity or opecity  Value for alpha varies from 0 to 1
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Random Data')
```

#### Bar Graph

```python
# This is Alphabetical vs Numerical Data
x=['a','b','c','d','e']
y=np.random.rand(5)
plt.bar(x,y)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Alphabetical vs Numerical Data')
```

```python
# This is Alphabetical vs Numerical Data in Horizontal Mode
x=['a','b','c','d','e']
y=np.random.rand(5)
plt.barh(x,y) #barh is used to show the Bar Graph in the Horizontal Mode 
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Alphabetical vs Numerical Data')
```

```python
x=np.linspace(0,10,200)
y=np.sin(x)
plt.figure(figsize=(6,2)) # To set the size of the Graph
plt.plot(x,y,'--r') # --r can be changed with --b,--y,--g or any other colour
plt.xlabel('This is X data')
plt.ylabel('This is Y data')
plt.title('This is Title')
plt.grid() # To Show the Gridlines
plt.show()
```

```python
x=np.random.rand(50)
y=np.random.rand(50)
sizes=200*np.random.rand(50) # *200 is just to increase the size of the dots you can change it depending on your need
colours=np.random.rand(50) # For different colours
plt.scatter(x,y,c='b',s=sizes,alpha=0.5) # Alpha is for intensity or opecity  Value for alpha varies from 0 to 1 , c='b' is to set the colour by yourself b=blue
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Random Data')
```

```python
x=np.random.rand(50)
y=np.random.rand(50)
sizes=200*np.random.rand(50) # *200 is just to increase the size of the dots you can change it depending on your need
colours=np.random.rand(50) # For different colours
plt.scatter(x,y,c='#FFA533',s=sizes,alpha=0.5,marker='x') # '#FFA533 is the HTML colour Code'
# Marker is to set the marker in the graph as here x is the marker and circle is by default not all but few keywords works as marker
plt.xlabel('X Axis') 
plt.ylabel('Y Axis')
plt.title('Random Data')
```

### Histogram

```python
# To plot a graph that tells te frequency of the repeated values in the Data
data=[1,2,3,4,5,1,2,3,6,7,8,1,2,3]
```

```python
plt.hist(data, color='y') # Here color is used for colour 
plt.xlabel('Data')
plt.ylabel('Number of Occurances')
plt.show
```

```python
x=[4,5,6,7,8,9]
y=[60,30,80,19,40,1]
colour=['red','blue','black','yellow','green','orange'] # Colours can be choosed in the form of list 
plt.scatter(x,y,c=colour)
plt.show()
```

### Representing in the 3D Coordinate as well

```python
# 3D-Scatter Plot
x=np.random.rand(50)
y=np.random.rand(50)
z=np.random.rand(50)

# For 3D 
fig=plt.figure() # Creating The Figure
ax=fig.add_subplot(projection='3d') # To tell the Projection must be 3D
ax.scatter(x,y,z)
plt.show()
```

```python
# 3D Line Plot
x=np.random.rand(50)
y=np.random.rand(50)
z=np.random.rand(50)

# For 3D 
fig=plt.figure() # Creating The Figure
ax=fig.add_subplot(projection='3d') # To tell the Projection must be 3D
ax.plot(x,y,z)
plt.show()
```
