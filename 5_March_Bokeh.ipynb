```python
from bokeh.plotting import figure ,output_file , show
from bokeh.sampledata.iris import flowers

output_file('test.html')

p=figure(title='test flower')
p.xaxis.axis_label='Petal Length'
p.yaxis.axis_label='Petal Width'
p.circle(flowers['petal_length'],flowers['petal_width'])
show(p)
```

Plotting The Line Graph

```python
from bokeh.plotting import figure ,output_file , show
from bokeh.sampledata.iris import flowers

output_file('Test_Line_Graph.html')

p=figure(title='test flower')
p.xaxis.axis_label='X Axis'
p.yaxis.axis_label='Y Axis'
p.line(flowers['petal_length'],flowers['petal_width'])
show(p)
```

```python
x=[2,3,4,5,6]
y=[3,4,5,6,7]
output_file=('line.html')
p=figure(title='line plot')
p.line(x,y)
show(p)
```

```python
x=[2,3,4,5,6]
y=[3,4,5,6,7]
output_file=('line.html')
p=figure(title='line plot')
p.scatter(x,y,fill_color='red') # To fill colour in the dots
show(p)
```

```python
x=[2,3,4,5,6]
y=[3,4,5,6,7]
output_file=('line.html')
p=figure(title='line plot')
p.scatter(x,y,fill_color='red',legend_label='red points') # To mark the points
show(p)
```

```python
x=[2,3,4,5,6]
y=[3,4,5,6,7]
output_file=('line.html')
p=figure(title='line plot')
p.scatter(x,y,fill_color='red',legend_label='red points',size=20) # Changing the size of circle
show(p)
```
