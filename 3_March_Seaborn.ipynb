```python
import seaborn as sns
```

```python
iris=sns.load_dataset('iris')
```

### Scatter Plot

```python
# Plotting the entire Scatter Plot
sns.scatterplot(iris)
```

```python
# Plotting the Sepal length vs Sepal Width Graph to check the relationship b/w Sepal Length and Sepal Width
sns.scatterplot(x=iris.sepal_length,y=iris.sepal_width)
```

```python
# Plotting the Sepal length vs Petal Length Graph to check the relationship b/w Sepal Length and Petal length
sns.scatterplot(x=iris.sepal_length,y=iris.petal_width)
```

```python
sns.scatterplot(x=iris.petal_length,y=iris.petal_width)
```

```python
sns.displot(iris['sepal_length']) # This is distribution Plot Displot
```

```python
sns.distplot(iris['sepal_length'])
```

```python
tips=sns.load_dataset('tips')
```

```python
tips
```

```python
# To check the relationship between the total bill and the tip 
sns.scatterplot(x=tips.total_bill,y=tips.tip)
```

```python
# The Graph outputs as Tip is Directly Proportional to Total Bill
```

```python
tips['smoker'].value_counts()
```

```python
# Show the smokers and the non smokers in the above scatterplot
sns.relplot(x=tips.total_bill,y=tips.tip,data=tips,hue='smoker') # Hue is used to represent the various categories present in that specific Data with the various colours
#relplot is for the relationship Plot 
```

```python
tips.head()
```

```python
# To check the size in the relationship with the total bill paid
sns.relplot(x=tips.total_bill,y=tips.tip,data=tips,hue='size') # Will generate the relationship between total bill and tip by seperating according to the size of the 
```

```python
# To check the size in the relationship with the total bill paid
sns.relplot(x=tips.total_bill,y=tips.tip,data=tips,hue='size',style='size') # Will generate the relationship between total bill and tip by seperating according to the size of the 
# style will let you represent the various categories of the data from the different different marks 
```

```python
# To check the size in the relationship with the total bill paid
sns.relplot(x=tips.total_bill,y=tips.tip,data=tips,style='size',hue='time') # Will generate the relationship between total bill and tip by seperating according to the size of the 
# style will let you represent the various categories of the data from the different different marks 
# Representing the 2 Grouping Together one with the Hue and other with the Style
```

### Categorical Plot

```python
sns.catplot(x='day',y='total_bill',data=tips)
```

```python
# to check who gives more tip Smoker or Non-Smoker
sns.catplot(x='smoker',y='tip',data=tips)
```

```python
sns.jointplot(x=tips.total_bill,y=tips.tip)
# Jointplot is showing two graphs at once Histogram and Scatterplot together
```

```python
sns.pairplot(tips)
# will generate all the possible numerical graphs and will be comparing each of them to every of them 
```
